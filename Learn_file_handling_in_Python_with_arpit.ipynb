### Note:
Q: What open file process does:
- Whenever you opens a file it means your file comes to the buffer mermory in RAM.

Q: Why closing is important:
- Closing is important because whenever you close the file after all your operations it removes from your buffer memory. Otherwise your RAM can be crashed if garbage collector is not removes it on time.
- It is important for the interview end.

# Using Context Manager (With)
- It's good idea to close a file after usage as it will free up the resources.
- If we don't close it, garbage collector would close it.
- 'with' keyword closes the file as soon as the usage is over.

```python
# with
with open('simple_two.txt', 'w') as f:
  f.write('Hello, I am a good person like my friend.\n');
```

```python
# try with read now
with open('simple_two.txt', 'r') as f:
  print(f.read());
```

```python
# Moving within a file -> 10 char then 10 char
with open('simple_two.txt', 'r') as f:
  print(f.read(10));
  print(f.read(10));
```

```python
# Benifit? -> to load the bog file in memory
big_L = ['Hello world ' for i in range(1000)];

with open('simple_three.txt', 'w') as f:
  f.writelines(big_L);
```

```python
# How read or load files in chunks
with open('simple_three.txt', 'r') as f:
  chunk_size = 10;
  while len(f.read(chunk_size)) > 0:
    print(f.read(chunk_size), end="**");
    f.read(chunk_size); # load the next chunk size in memory.
```

```python
# seek and tell function
with open('simple_three.txt', 'r') as f:
  f.seek(10); ## Means curser positioned to the 10th char
  print(f.read(10)); ## read next 10 chars
  print(f.tell()); ## returns a postion where curser points.

  f.seek(0);
  print(f.read(10));
  print(f.tell());
```

```python
from io import SEEK_END
# Test: How to point end of th e file.
with open('simple_three.txt', 'r') as f:
  f.seek(0, SEEK_END);
  print(f.tell());
```

```python
# seek during write -> Important to know
with open('simple.txt', 'w') as f:
  f.write('Hello');
  f.seek(0);
  f.write('X'); ## Overwrite -> observe the chnages in the file.
```
