# Using Context Manager (With)
- It's good idea to close a file after usage as it will free up the resources.
- If we don't close it, garbage collector would close it.
- 'with' keyword closes the file as soon as the usage is over.

```python
# with
with open('simple_two.txt', 'w') as f:
  f.write('Hello, I am a good person like my friend.\n');
```

```python
# Moving within a file -> 10 char then 10 char
with open('simple_two.txt', 'r') as f:
  print(f.read(10));
  print(f.read(10));
```

```python
# Benifit? -> to load the log file in memory
big_L = ['Hello world ' for i in range(1000)];

with open('simple_three.txt', 'w') as f:
  f.writelines(big_L);
```

```python
# How read or load files in chunks
with open('simple_three.txt', 'r') as f:
  chunk_size = 10;
  while len(f.read(chunk_size)) > 0:
    print(f.read(chunk_size), end="**");
    f.read(chunk_size); # load the next chunk size in memory.
```

```python
# seek and tell function
with open('simple_three.txt', 'r') as f:
  f.seek(10); ## Means curser positioned to the 10th char
  print(f.read(10)); ## read next 10 chars
  print(f.tell()); ## returns a postion where curser points.

  f.seek(0);
  print(f.read(10));
  print(f.tell());
```

```python
from io import SEEK_END
# Test: How to point end of th e file.
with open('simple_three.txt', 'r') as f:
  f.seek(0, SEEK_END);
  print(f.tell());
```

```python
# seek during write -> Important to know
with open('simple.txt', 'w') as f:
  f.write('Hello');
  f.seek(0);
  f.write('X'); ## Overwrite -> observe the chnages in the file.
```

### Serilalization and Deserialiazation

```python
class Person:
  def __init__(self, fname, lname, age, gender):
    self.fname = fname;
    self.lname = lname;
    self.age = age;
    self.gender = gender;
# Format to printed in
# -> Arpit Jain age -> 23 gender -> male
```

```python
arpit = Person('Arpit', 'Jain', 23, 'male');
```

```python
import json

with open('demo.json', 'w') as f:
  json.dump(arpit.__dict__, f); # Person object to dictionary
```

```python
with open('demo.json', 'r') as f:
  p = json.load(f);
  print(p);
  print(type(p));
```

```python
# As a string

def show_object(person):
  return f'{person.fname} {person.lname} age -> {person.age} gender -> {person.gender}';

with open('demo.json', 'w') as f:
  json.dump(arpit, f, default=show_object);
#
```

```python
with open('demo.json', 'r') as f:
  p = json.load(f);
  print(p);
  print(type(p));
```

## Pickling

- `Pickling` is the process whereby a Python object hierarchy is converted into a byte stream
- `Unpickling` is the inverse operation, where a byte stream (from a binary file or file like object) is converted back into a object hierarchy.

```python
class Person:

  def __init__(self, name, age):
    self.name = name;
    self.age = age;

  def display(self):
    print(f'Name: {self.name} Age: {self.age}');
```

```python
akash = Person('Akash', 23);
```

```python
import pickle
with open('person.pkl', 'wb') as f:
  pickle.dump(akash, f);
```

```python
import pickle
with open('person.pkl', 'rb') as f:
  p = pickle.load(f);
  p.display(); ## See we can access object and their properties via file
```
