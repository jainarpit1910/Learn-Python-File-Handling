Here's the explanation of Python file handling concepts, focusing on simplicity and essential details for beginners:\

## Types of data used for I/O (Input/Output)
- Text: '12345' as a sequence of unicode.
- Binary: 12345 as a sequence of binary equivalent.

There are two types to deal with:
- Text: All programming files and text files.
- Binary: All images, musics, videos, and exe files.

#### How files I/O is done in the most programming language:
- Open a file
- Read/Write file
- Close the file

### Note:
Q: What file open process done:
- Whenever you opens a files it means your file comes to the Buffer mermory in RAM.

Q: Why closing is important:
- Closing is important because whenever you close the files after all your operation it removes form your Buffer memory. Otherwise you RAM can be crashed if garbage collector is not remove it on time.
- It is important for the Interview end.


```python
## Basic code for the write
f = open('simple.txt', 'w'); ## open a file. (file_path, mode of open)
# It create a new file if its is not present with name simple.txt
f.write('Hello, I am a Developer.');
f.close();
## Sice file is closed hence this will not work
# f.write('Hello, I am a Developer2.');  ## uncomment and check
```

```python
# Write multiple lines in string
f = open('simple_one.txt', 'w');
f.write('Hello, I am a Developer.\n');
f.write('Hello, I am a good person like my friend.\n');
f.close();
```

```python
## Case 1: If file has already have data
f = open('simple.txt', 'w');
f.write('Hello, someone hacked you'); ## Overwrite the data - It is important that please do not try with important files.
# It can be crash your system and you can loose some important data
f.close();
```

```python
# Introduction of append mode.
f = open('simple.txt', 'a');
f.write('\nHello, I am a good person like my friend.\n');
f.close();
```

```python
# Write lines writelines()
L = ['Hello\n', 'Hope you are doing well\n', 'We are learning python\n'];
f = open('simple_one.txt', 'w');
f.writelines(L);
f.close();
```

```python
# Read form files -> using read()
f = open('simple_one.txt', 'r');
s = f.read();
f.close();
print(s);
```

```python
# if you want to print 20 chars.
f = open('simple_one.txt', 'r');
s = f.read(20);
f.close();
print(s);
```

```python
# Read form files -> using readline()
# f = open('simple_one.txt', 'r');
f = open('simple_one.txt', 'r');
s1 = f.readline();
s2 = f.readline();
f.close();
print(s1, end='');
print(s2, end='');
```

```python
# Reading entire file using readline
f = open('simple_one.txt', 'r');

while True:
  s = f.readline();
  if s == '':
    break;
  else:
    print(s, end='');
f.close();
```
